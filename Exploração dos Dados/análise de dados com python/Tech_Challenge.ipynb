```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# A função "melt" vai reduzir o conjunto de dados uma relação de informações considerada primordiais. O "id_vars=[]" é "Qual a informação que irá identificar as variáveis"
# Nesse caso é a coluna de datas que está com o nome "index". Depois identificamos onde estão os valores que queremos puxas, com o "value_vars="
#mensal_aberto = mensal.reset_index().melt(id_vars=["index"], value_vars=mensal.columns)
#mensal_aberto.columns=['dia_mes_ano','uf','gasto']
#mensal_aberto.head()
```

```python
prod_vinho = prod_vinho.reset_index().melt(id_vars=['produto'],value_vars=prod_vinho.columns)
prod_vinho.columns=['produto','ano','volume']
prod_vinho.head()
```

```python
prod_vinho['produto'].unique()
```

```python
plt.figure(figsize = (10,6))
axis = sns.lineplot(data = prod_vinho.query("(produto == 'Tinto') | (produto == 'Branco') | (produto == 'Rosado') | (produto == 'VINHO FINO DE MESA (VINÍFERA)')"), x = 'ano', y='volume',hue = 'produto')
plt.ylim(0,prod_vinho['volume'].max()*1.1)
plt.grid(linestyle="--")
plt.show()
```

##**Mesclando as tabelas**##

```python
exp_vinho_vol_ano_grupo.head()
```

```python
prod_vinho_ano = prod_vinho.query("(produto == 'VINHO DE MESA') | (produto == 'VINHO FINO DE MESA (VINÍFERA)') | (produto == 'DERIVADOS') | (produto == 'SUCO')").groupby('ano').sum("volume")
vinho_prod_x_exp = pd.merge(prod_vinho_ano,exp_vinho_vol_ano_grupo, on ='ano')
vinho_prod_x_exp.head()
```

```python
vinho_prod_x_exp = vinho_prod_x_exp.rename(columns={'volume_x':'Volume Produzido','volume_y':'Volume Exportado'})
vinho_prod_x_exp.head()
```

```python
vinho_prod_x_exp['porct_uso'] = (vinho_prod_x_exp['Volume Exportado'] / vinho_prod_x_exp['Volume Produzido'])*100
vinho_prod_x_exp.head()
```

```python
plt.figure(figsize = (10,6))
axis = sns.lineplot(data = vinho_prod_x_exp, x = vinho_prod_x_exp.index, y='porct_uso')
plt.ylim(0,vinho_prod_x_exp['porct_uso'].max()*1.1)
plt.grid(linestyle="--")
plt.show()
```

```python
vinho_prod_x_exp['porct_disponivel'] = 100-vinho_prod_x_exp['porct_uso']
plt.figure(figsize = (10,6))
axis = sns.lineplot(data = vinho_prod_x_exp, x = vinho_prod_x_exp.index, y='porct_disponivel')
plt.ylim(80,vinho_prod_x_exp['porct_disponivel'].max()*1.1)
plt.grid(linestyle="--")
plt.show()
```
