```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
```

```python
valores = pd.read_excel('/content/valoresexportacao.xlsx')
quantidade = pd.read_excel('/content/quantidadeexportacao.xlsx')
```

```python
valores
```

```python
quantidade
```

```python
quantidade.drop('Id', inplace=True, axis=1)
```

```python
valores.drop('Id', inplace=True, axis=1)
```

```python
quantidade.set_index('País', inplace=True)
valores.set_index('País', inplace=True)
```

```python
quantidade['Total_qtde'] = quantidade.sum(axis=1)
valores['Total'] = valores.sum(axis=1)
```

```python
quantidade.sort_values('Total_qtde', ascending=False, inplace=True)
valores.sort_values('Total', ascending=False, inplace=True)
```

```python
quantidade['Total_qtde'].head(10).plot(kind='bar', figsize=(10, 6))
plt.title('Quantidade de exportação total por país')
plt.xlabel('País')
plt.ylabel('Quantidade')
plt.show()
```

```python
valores['Total'].head(10).plot(kind='bar', figsize=(10, 6), color='red')
plt.title('Quantidade de exportação por país')
plt.xlabel('País')
plt.ylabel('Quantidade')
plt.show()
```

```python
quantidade_total = quantidade
valores_total = pd.read_excel('/content/valoresexportacao_total.xlsx')
```

```python
valores_total.drop('Id', inplace=True, axis=1)
```

```python
valores_total.set_index('País', inplace=True)
```

```python
quantidade_total.head(10)
quantidade_total.drop(columns=['2007','2008','2009','2010','2011','2012','2013','2014','2015','2016','2017','2018','2019','2020','2021','2022','2023'], axis=1, inplace=True)
```

```python
quantidade_total.head(10)
```

```python
valores_total.head(10)
```

```python
valor_e_quantidade = quantidade_total.join(valores_total)
```

```python
valor_e_quantidade
```

```python
axis = valor_e_quantidade.head(10).plot(kind='bar', figsize=(10, 6))

axis.set_xlabel("Países")
axis.set_ylabel("Valor em Milhões QTDE/VALOR")
axis.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2F}"))
axis.tick_params(axis='x', labelrotation=45 )
axis.grid(True, linestyle='-.')

plt.title('Quantidade/valor total de exportação por país - VINHO')
plt.show()
```
